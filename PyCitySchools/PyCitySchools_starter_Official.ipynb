```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

# District Summary

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

```python
#Calculate the total number of schools
total_number_schools = len(school_data_complete["school_name"].unique())
total_number_schools
```

```python
#Calculate the total number of students
total_number_students = student_data["Student ID"].count()
total_number_students
```

```python
#Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculate the average math score 
avg_math = student_data["math_score"].mean()
avg_math
```

```python
#Calculate the average reading score
avg_reading = student_data["reading_score"].mean()
avg_reading
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_passing_df = school_data_complete.loc[school_data_complete["math_score"] >= 70]
math_passing_number = math_passing_df["Student ID"].count() 
math_passing_pct = (math_passing_number / total_number_students) * 100

math_passing_pct 
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
reading_passing_df = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
reading_passing_number = reading_passing_df["Student ID"].count() 
reading_passing_pct = (reading_passing_number / total_number_students) * 100

reading_passing_pct 
```

```python
# Calculate the percentage of students who passed math **and** reading (% Overall Passing)
overall_passing_df = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]
overall_passing = overall_passing_df['Student ID'].count()/total_number_students*100
overall_passing
```

```python
# Create a dataframe to hold the above results
results_df = pd.DataFrame({
           "Total Number of Schools":total_number_schools, 
           "Total Number of Students":f"{total_number_students:,}",
           "Total Budget":f"${total_budget:,.2f}",
           "Average Math Score": round(avg_math,2), 
           "Average Reading Score": round(avg_reading,2),
           "Percent Students Passing Math":round(math_passing_pct,2),
           "Percent Students Passing Reading":round(reading_passing_pct,2),
           "Overall Passing Rate":round(overall_passing,2)},
           index=[0])
results_df
```

## Math Scores by Grade

* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Create a pandas series for each grade. Hint: use a conditional statement.
#Group each series by school
ninth_avg_score = student_data.loc[student_data['grade'] == '9th']
ninth_avg_score_school = ninth_avg_score.groupby('school_name')["math_score"].mean()
ninth_avg_score_school

tenth_avg_score = student_data.loc[student_data['grade'] == '10th']
tenth_avg_score_school = tenth_avg_score.groupby('school_name')["math_score"].mean()
tenth_avg_score_school

eleventh_avg_score = student_data.loc[student_data['grade'] == '11th']
eleventh_avg_score_school = eleventh_avg_score.groupby('school_name')["math_score"].mean()
eleventh_avg_score_school

twelfth_avg_score = student_data.loc[student_data['grade'] == '12th']
twelfth_avg_score_school = twelfth_avg_score.groupby('school_name')["math_score"].mean()
twelfth_avg_score_school

#Combine the series into a dataframe
avg_math_score_school_df = pd.DataFrame({
    "9th": ninth_avg_score_school,
    "10th": tenth_avg_score_school,
    "11th": eleventh_avg_score_school,
    "12th": twelfth_avg_score_school
})

avg_math_score_school_df

#Optional: give the displayed data cleaner formatting
```

## Scores by School Spending

  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins in which to place values based upon IMDB vote count
bins = [0, 585, 630, 645, 680]

# Create labels for these bins
group_money_range= ["<$585", "$585-630", "$630-645", "$645-680"]

# Slice the data and place it into bins
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_money_range)

# Create a GroupBy object based upon "Spending Ranges (Per Student)"
spending_group = school_summary_df.groupby('Spending Ranges (Per Student)')
spending_group

# Find the average that falls into each bin
spending_group_avg = spending_group.mean()[['Average Math Score', 'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]
spending_group_avg
```
